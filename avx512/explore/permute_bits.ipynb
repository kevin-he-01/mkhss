```python
from matplotlib import pyplot as plt
import random
```

```python
def experiment(n: int, t: int, ntrials: int = 1000, plot: bool = True) -> None:
    lows = [n] * n
    highs = [0] * n
    binstring = [1] * t + [0] * (n - t)
    for _ in range(ntrials):
        sum_ = 0
        random.shuffle(binstring)
        for i in range(n):
            sum_ += binstring[i]
            lows[i] = min(lows[i], sum_)
            highs[i] = max(highs[i], sum_)
    # Compute the number of nodes in the optimized branching program
    num_nodes = 0
    for i in range(n):
        num_nodes += highs[i] - lows[i]
    num_nodes_unoptimized = n * t - t * t
    print(f"n = {n}, t = {t}, ntrials = {ntrials}, improvement = {num_nodes_unoptimized / num_nodes}")
    if plot:
        plt.plot(range(n), lows, label='min')
        plt.plot(range(n), highs, label='max')
        plt.title(f"n = {n}, t = {t}, ntrials = {ntrials}")
        plt.show()
```

```python
for n in [32, 64]:
    for ntrials in [10, 100, 1000, 10000, 100000, 1000000]:
        experiment(n, n // 3, ntrials=ntrials, plot=False)
```

```python
for n in [128, 256, 512, 1024]:
    for ntrials in [10, 100, 1000, 10000]:
        experiment(n, n // 3, ntrials=ntrials, plot=False)
```

```python
for n in [128, 256, 512, 1024, 2048]:
    experiment(n, n // 3, ntrials=1000)
```
